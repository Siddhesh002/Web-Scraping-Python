## Part 1

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from time import sleep
import pandas as pd
```

```python
path = "C://chromedriver.exe"
```

```python
caps = DesiredCapabilities().CHROME
caps["pageLoadStrategy"] = "eager"

browser = webdriver.Chrome(desired_capabilities=caps,executable_path=path)
```

```python
no_of_pages = 5
```

```python
products_name = []
all_urls = []


for i in range(1,no_of_pages):
 
    print("Scraping page",i)
    
    product = browser.find_elements(By.XPATH,"//span[@class='a-size-medium a-color-base a-text-normal']")
    url = browser.find_elements(By.XPATH,"//span[@class='a-size-medium a-color-base a-text-normal']/..")
    
    for p in product:
      products_name.append(p.text)
    
    urls = [u.get_attribute('href') for u in url]
    for u in urls:
      all_urls.append(u)  
    
    browser.get(f"https://www.amazon.in/s?k=bags&page={i}")
    sleep(2)
 
df = pd.DataFrame()

df['Product_Url'] = all_urls
df['Product_Name'] = products_name
```

```python
df.dropna(inplace=True)
```

```python
df['Product Price'] = ''
df['Rating'] = ''
df['Number of reviews'] = ''
```

```python
for url in df['Product_Url']:
  try :
    browser.get(url)
    price = browser.find_element(By.XPATH,"//span[@class='a-price aok-align-center reinventPricePriceToPayMargin priceToPay']")
    rating = browser.find_element(By.XPATH,"//span[@data-hook='rating-out-of-text']")
    nor = browser.find_element(By.XPATH,"//div[@class='a-row a-spacing-medium averageStarRatingNumerical']")
    df.loc[df['Product_Url'] == url, 'Product Price'] = price.text
    df.loc[df['Product_Url'] == url, 'Rating'] = rating.text
    df.loc[df['Product_Url'] == url, 'Number of reviews'] = nor.text

  except:
    pass
```

```python
import numpy as np
df.replace('',np.NaN,inplace=True)
```

```python
df.dropna(inplace=True)
```

## Part 2

```python
df['Description'] = ''
df['ASIN'] = ''
df['Product Description'] = ''
df['Manufacturer'] = ''
```

```python
for url in df['Product_Url']:
  try :
    browser.get(url)
    description = browser.find_element(By.XPATH,"//ul[@class='a-unordered-list a-vertical a-spacing-mini']")
    prod_description = browser.find_element(By.XPATH,"//ul[@class='a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list']")
    prod_description_dict = dict(item.split(' : ') for item in prod_description.text.split('\n'))
    df.loc[df['Product_Url'] == url, 'Description'] = description.text
    df.loc[df['Product_Url'] == url, 'ASIN'] = prod_description_dict['ASIN']
    df.loc[df['Product_Url'] == url, 'Product Description'] = prod_description.text
    df.loc[df['Product_Url'] == url, 'Manufacturer'] = prod_description_dict['Manufacturer']
  except:
    pass
```

```python
df
```

```python
import numpy as np
df.replace('',np.NaN,inplace=True)
```

```python
df.dropna(inplace=True)
```
